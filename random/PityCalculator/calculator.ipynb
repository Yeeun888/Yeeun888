```python
#Imports so the library doesn't have to be loaded in
import random
import matplotlib.pyplot as plt
```

```python
#Pity Calculator

#Set Starting Variables
lastFiftyFiftyWon = True
constelationGoal = 3

initialState = lastFiftyFiftyWon
def checkPity(pity):
    if(pity == 90):
        return True

    return random.random() < (pityProbability[pity] / 100)

def goodFiveStar():
    global lastFiftyFiftyWon

    if(lastFiftyFiftyWon == False):
        lastFiftyFiftyWon = True
        return True

    #Code below only executes if lastFiftyFiftyWon is True
    if(random.random() > 0.5):
        return True
    else:
        #User lost the 50/50, therefore it is false, such that the next one is true
        lastFiftyFiftyWon = False
        return False
```

```python
simulationIterations = 1000000
results = []

for i in range(0, simulationIterations):
    #For a single wishing session with infinite tries
    totalRolls = 0
    iterationCurrentPity = 0
    currentPity = 0
    constelationsWon = 0

    while(True):
        if(checkPity(currentPity)):
            #Code if get 5 star here
            if(goodFiveStar()):
                #We won the 50/50
                constelationsWon += 1
                currentPity = 0

            else:
                #We lost the 50/50
                currentPity = 0
                
        currentPity += 1
        iterationCurrentPity += 1

        if(constelationsWon == constelationGoal):
            results.append(iterationCurrentPity)
            lastFiftyFiftyWon = initialState
            break
```

```python
plt.hist(results, bins = 100)

plt.xlabel("Number of Rolls Required")
plt.ylabel("Total Samples (out of 1m simulations)")
plt.title("Rolls to get %i copies of character" % (constelationGoal))
plt.text(y=50000, x = 0, s="Previous 50/50 Win %s" % (initialState))
```
